```python
import sqlite3
import pandas as pd

class dataBase():
    """
    Simple CM for sqlite3 databases. Commits everything at exit.
    """
    def __init__(self, db_location):
        self.db_location = 'pension2014.db'

    def __enter__(self):
        self.conn = sqlite3.connect(self.db_location)
        self.cursor = self.conn.cursor()
        return self

    def __exit__(self, exc_class, exc, traceback):
        self.conn.close()
```

```python
#Read the csv file and taking our related details
tb_values = pd.read_csv('data/pension_2014.csv')
geotb = tb_values.loc[:, ['GEO.id', 'GEO.display-label']].drop_duplicates()
govtypetb = tb_values.loc[:, ['GOVTYPE.id', 'GOVTYPE.display-label']].drop_duplicates()
penitemtb = tb_values.loc[:, ['PENITEM.id', 'PENITEM.display-label']].drop_duplicates()
penvaluetb = tb_values.loc[:, ['GEO.id', 'GOVTYPE.id','PENITEM.id', 'AMOUNT' ]].drop_duplicates()
```

```python
#Drop the tables if needed
with dataBase('pension.db') as sqls:    
        sqls.cursor.execute('Drop table if exists geotb;')
        sqls.cursor.execute('Drop table if exists govtypetb;')
        sqls.cursor.execute('Drop table if exists penitemtb;')
        sqls.cursor.execute('Drop table if exists penvaluetb;')
```

```python
#creating tables
with dataBase('pension.db') as sqls: 
    #Create geotb table with required details - uncomment if needed 
    sqls.cursor.execute('create table geotb([GEO.id] text primary key, [GEO.display-label] text);')
    geotb.to_sql('geotb', sqls.conn, if_exists='append', index = False)
   
      #Create govtypetb table with required details
    sqls.cursor.execute('create table govtypetb([GOVTYPE.id] integer primary key, [GOVTYPE.display-label] text);')
    govtypetb.to_sql('govtypetb', sqls.conn, if_exists='append', index = False)
    
      #Create penitemtb table with required details
    sqls.cursor.execute('create table penitemtb([PENITEM.id] integer primary key, [PENITEM.display-label] text);')
    penitemtb.to_sql('penitemtb', sqls.conn, if_exists='append', index = False)
    
      #Create penvaluetb table with required details
    sqls.cursor.execute('create table penvaluetb([GEO.id] text, [GOVTYPE.id] integer, [PENITEM.id] integer, [AMOUNT] integer, primary key([GEO.id], [GOVTYPE.id], [PENITEM.id]));')
    penvaluetb.to_sql('penvaluetb', sqls.conn, if_exists='append', index = False)
```

```python
#getting the records in dataframe format from geotb table
with dataBase('pension.db') as sqls:
    geotb_df = pd.read_sql('SELECT [GEO.id] as geoid,[GEO.display-label] as geoname FROM geotb;', con = sqls.conn)
    display(geotb_df)
```

```python
#getting the records in dataframe format from govtype table
with dataBase('pension.db') as sqls:
    govtypetb_df = pd.read_sql('SELECT [GOVTYPE.id] as govtypeid,[GOVTYPE.display-label] as govtypelabel FROM govtypetb;', con = sqls.conn)
    display(govtypetb_df)
```

```python
#getting the records in dataframe format from penitemtb table
with dataBase('pension.db') as sqls:
    penitemtb_df = pd.read_sql('SELECT [PENITEM.id] as penitemid,[PENITEM.display-label] as penitemlabel FROM penitemtb;', con = sqls.conn)
    display(penitemtb_df)
```

```python
#getting the records in dataframe format from penvaluetb table
with dataBase('pension.db') as sqls:
    penvaluetb_df = pd.read_sql('SELECT [GEO.id] as geoid,[GOVTYPE.id] as govtypeid, [PENITEM.id] as  penitemid, [AMOUNT] as amount FROM penvaluetb;', con = sqls.conn)
    display(penvaluetb_df)
```

```python
#retrieve data from all the tables and making the same format as pension_2014.csv
with dataBase('pension.db') as sqls:
    retrieve_df = pd.read_sql('SELECT g.[GEO.id],g.[GEO.display-label], g2.[GOVTYPE.id], g2.[GOVTYPE.display-label], g4.[PENITEM.id], g4.[PENITEM.display-label], g1.[AMOUNT]  FROM geotb g inner join penvaluetb g1 on g.[GEO.id]=g1.[GEO.id] inner join govtypetb g2 on g1.[GOVTYPE.id]=g2.[GOVTYPE.id] inner join penitemtb g4 on g1.[PENITEM.id] = g4.[PENITEM.id];', con = sqls.conn)
    display(retrieve_df)
```

```python
#Dealing with ratio from Employee Contributions to Total Contributions
with dataBase('pension.db') as sqls:
    ratio_df = pd.read_sql('SELECT [GEO.display-label], [GOVTYPE.display-label], max(case when [PENITEM.display-label]=\'   Total Contributions\' then [AMOUNT] end) as Total, max(case when [PENITEM.display-label]=\'   Employee contributions\' then [AMOUNT] end) as Employee, round(cast(max(case when [PENITEM.display-label]=\'   Employee contributions\' then [AMOUNT] end) as REAL)/ max(case when [PENITEM.display-label]=\'   Total Contributions\' then [AMOUNT] end), 1) as Ratio FROM geotb g inner join penvaluetb g1 on g.[GEO.id]=g1.[GEO.id] inner join govtypetb g2 on g1.[GOVTYPE.id]=g2.[GOVTYPE.id] inner join penitemtb g4 on g1.[PENITEM.id] = g4.[PENITEM.id] where [GOVTYPE.display-label]=\'State and Local \' and [PENITEM.display-label] in (\'   Total Contributions\', \'   Employee contributions\') group by [GEO.display-label], [GOVTYPE.display-label] order by Ratio desc, Total desc;', con = sqls.conn)
    #Plotting the data in bar plot - state as x-axis and EM/Total as y-axis
    ax = ratio_df.plot.bar(x='GEO.display-label', y='Ratio', title="Employee/Total ratio",figsize=(20, 6))
    # ax.set_ylim(0.5, 1.0)
    ax.set_xlabel("state")
    ax.set_ylabel("EM/Total")
```
